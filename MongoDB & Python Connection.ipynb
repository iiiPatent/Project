```python
from pymongo import MongoClient


# Connecting Database
client = MongoClient("localhost:27017")


# Choose database
db = client['jeffSheng']




user_id = '595620944'

f = open('C:/Users/BigData/1st_html/Post_Content//{}.txt'.format(user_id),'r')
text = f.readlines()
f.close()

# Building a json file
 
post = []

for ele in text:
    post.append({'time' : ele.strip().split(',')[0] , 'text':str(ele.strip().split(',')[1])})
json = {'user_id':user_id , 'post':post }


# db.collection.function()

db.fb_posts.insert(json)
```
