```python
from collections import Counter
from sklearn import  cluster
import numpy as np
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage
from sklearn.cluster import hierarchical
```

```python
model.init_sims()
```

```python
model.syn0norm
```

```python
kmean = cluster.KMeans(n_clusters=10)
result = kmean.fit(model.syn0norm)
```

```python
Counter(result.labels_)
```

```python
label = 4
print " ".join(np.array(model.index2word)[result.labels_ == label])
```

```python
from sklearn.cluster import hierarchical
```

```python
index = result.labels_ == label
dist = pdist(np.array(model.syn0norm)[index])
m = linkage(dist, method='average')
h = hierarchical.hierarchy.fcluster(m, 0.5*dist.max(), 'distance')
```
